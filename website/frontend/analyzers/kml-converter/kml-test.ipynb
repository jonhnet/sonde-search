```python
import requests
import pandas as pd

df = pd.DataFrame(requests.get('https://api.v2.sondehub.org/sonde/V1854526').json())
df['datetime'] = pd.to_datetime(df['datetime'])
df = df.set_index('datetime')[['lon', 'lat', 'alt']]
display(df.loc[df['alt'] > 30000])
```

```python

s = pd.concat([
    df.head(1),
    df.resample('1 min', label='right').last(),
    df.loc[df['alt'].idxmax()].head(1),
]).dropna().sort_index()
pd.set_option('display.min_rows', 150)
pd.set_option('display.max_rows', 500)

display(s)
```

```python
coords = list(s.itertuples(index=False, name=None))
display(coords)
```

```python
import simplekml
kml = simplekml.Kml()
kml.document.name='test'
linestring = kml.newlinestring(name='Serial')
linestring.coords = coords
linestring.altitudemode = simplekml.AltitudeMode.absolute
linestring.extrude = 1
linestring.style.linestyle.color = simplekml.Color.red
linestring.style.linestyle.width = 5
kml.save('test.kml')
```

```python
import xml.etree.ElementTree as ET
root = ET.fromstring(kml.kml())
root
```
